### Read Data

```python
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NLP Project/Final/data/final_train.csv')
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NLP Project/Final/data/final_test.csv')
```

```python
test.fillna('', inplace=True)
train.fillna('', inplace=True)
```

## Create vector representation

```python
text_blocks_train  = list(train['text'])
text_blocks_test  = list(test['text'])
```

```python
y_train = list(train['sub'])
y_test  = list(test['sub'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
vectorizer = CountVectorizer(min_df=0, lowercase=False)
```

```python
vectorizer.fit(text_blocks_train)
```

```python
X_train = vectorizer.transform(text_blocks_train)
X_test = vectorizer.transform(text_blocks_test)
```

### Create the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(multi_class='multinomial', solver='lbfgs')
```

```python
classifier.fit(X_train, y_train)
```

```python
score = classifier.score(X_test, y_test)
print("Accuracy:", score)
```

### Overview of the model

```python
for i, test_string in enumerate(list(test['text'])):
    text = vectorizer.transform([test_string])
    print("Predicted:", classifier.predict(text), "Effektiv:", list(test['sub'])[i])
```
